```python
# Group Project

'''  Title: Electricity Billing System
     Subject Name: Python
     Date: 5-April-24 '''

import matplotlib.pyplot as plt
import random


class ElectricityBill:
    FIXED_CHARGES = 50.0  # Adjust the fixed charges as needed
    GOVERNMENT_DUTY_RATE = 0.15  # 15% government duty
    
    def __init__(self, customer_name, units_consumed):
        self.customer_name = customer_name
        self.units_consumed = units_consumed
        self.total_bill = self.calculate_bill()

    def calculate_bill(self):
        # Calculate the electricity bill based on your pricing logic
        # You can modify this calculation as per your requirements
        rate_per_unit = 5.0  # Adjust the rate as needed

        # Calculate the basic bill
        basic_bill = self.units_consumed * rate_per_unit

        # Add fixed charges
        total_bill_before_duty = basic_bill + ElectricityBill.FIXED_CHARGES

        # Apply government duty
        government_duty = total_bill_before_duty * ElectricityBill.GOVERNMENT_DUTY_RATE

        # Calculate the total bill
        total_bill = total_bill_before_duty + government_duty

        return total_bill

    def print_bill(self):
        print(f"Customer Name: {self.customer_name}")
        print(f"Units Consumed: {self.units_consumed}")
        print(f"Total Bill: â‚¹{self.total_bill}")
    
class RandomDataGenerator:
    @staticmethod
    def generate_random_data(years):
        # Generate random data for each year
        data = [random.randint(50, 200) for _ in range(years)]
        return data


class GraphPlotter:
    @staticmethod
    def plot_random_graph(years, data):
        # Plotting
        plt.plot(range(1, years + 1), data, marker='o', linestyle='-')
        plt.title('Electricity Consumed')
        plt.xlabel('Year')
        plt.ylabel('Units')
        plt.grid(True)
        plt.show()
        
class User:
    def __init__(self, username, password):
        self.username = username
        self.password = password

    def get_username(self):
        return self.username

    def get_password(self):
        return self.password
        
class UserManagement:
    registered_users = []

    @staticmethod
    def login():
        print("\n----- User Login -----")
        username = input("Enter username: ")
        password = input("Enter password: ")

        user = UserManagement.authenticate_user(username, password)

        if user:
            print("Login successful!")
            return user
        else:
            print("Invalid username or password. Please try again.")
            return None

    @staticmethod
    def authenticate_user(username, password):
        for user in UserManagement.registered_users:
            if user.get_username() == username and user.get_password() == password:
                return user
        return None

    @staticmethod
    def register():
        print("\n----- User Registration -----")
        new_username = input("Enter a new username: ")
        new_password = input("Enter a new password: ")

        UserManagement.registered_users.append(User(new_username, new_password))
        print("Registration successful. You can now log in.")


def main():
    try:
        print("===============================================================================")
        print("                   Welcome to the Electricity Billing System!                  ")
        print("===============================================================================")

        while True:
            print("\nOptions:")
            print("1. Register")
            print("2. Login")
            print("3. Exit")

            choice = input("Enter your choice: ")

            if choice == '1':
                UserManagement.register()
            elif choice == '2':
                user = UserManagement.login()
                if user:
                    break
            elif choice == '3':
                print("Exiting...")
                return
            else:
                print("Invalid choice. Please enter '1', '2', or '3'.")
        # Take user input
        customer_name = input("Enter Customer Name: ")
        units_consumed = int(input("Enter Units Consumed: "))

        # Create an ElectricityBill object
        bill = ElectricityBill(customer_name, units_consumed)

        # Print the bill details
        bill.print_bill()

        while True:
            years = int(input("Enter the number of years for the graph: "))
            if years <= 0:
                print("Number of years must be greater than zero.")
            else:
                break
        # Generate random data
        random_data = RandomDataGenerator.generate_random_data(years)

        # Plot random graph
        GraphPlotter.plot_random_graph(years, random_data)
        
        print("Returning to the main menu...")
        main()

    except Exception as e:
        print(f"An error occurred: {e}")

if __name__ == "__main__":
    main()
```

```python
# Individual Project

'''  Title: Car Rental System
     Enrollment: 22002171210123
     Subject Name: Python
     Date: 5-April-24 '''

import matplotlib.pyplot as plt
import random


class Car:
    def __init__(self, name, category, daily_rate, seating_capacity, fuel_type):
        self.name = name
        self.category = category
        self.daily_rate = daily_rate
        self.seating_capacity = seating_capacity
        self.fuel_type = fuel_type

    def get_name(self):
        return self.name

    def get_category(self):
        return self.category

    def get_daily_rate(self):
        return self.daily_rate

    def get_seating_capacity(self):
        return self.seating_capacity

    def get_fuel_type(self):
        return self.fuel_type


class User:
    def __init__(self, username, password):
        self.username = username
        self.password = password

    def get_username(self):
        return self.username

    def get_password(self):
        return self.password


class Reservation:
    def __init__(self, user, car, reservation_date, rental_days):
        self.user = user
        self.car = car
        self.reservation_date = reservation_date
        self.rental_days = rental_days
        self.total_amount = rental_days * car.get_daily_rate()

    def get_user(self):
        return self.user

    def get_car(self):
        return self.car

    def get_reservation_date(self):
        return self.reservation_date

    def get_rental_days(self):
        return self.rental_days

    def get_total_amount(self):
        return self.total_amount


class CarRentalSystem:
    users = []
    cars = []
    reservations = []
    logged_in_user = None

    @staticmethod
    def main():
        CarRentalSystem.initialize_data()

        while True:
            print("===============================================================================")
            print("                        Welcome to the Car Rental System                       ")
            print("===============================================================================")

            print("1. Register")
            print("2. Login")
            print("3. Exit")
            choice = int(input("Select an option: "))

            if choice == 2:
                CarRentalSystem.login()
            elif choice == 1:
                CarRentalSystem.register()
            elif choice == 3:
                print("Exiting the system.")
                break
            else:
                print("Invalid choice. Please try again.")

    @staticmethod
    def initialize_data():
        CarRentalSystem.cars = [
            Car("Toyota Fortuner", "SUV", 700.0, 7, "Petrol"),
            Car("Toyota Innova", "SUV", 650.0, 7, "Diesel"),
            Car("BMW X1", "Luxury", 600.0, 5, "Petrol"),
            Car("BMW X3", "Luxury", 220.0, 5, "Diesel"),
            Car("Audi Q7", "Luxury", 1300.0, 7, "Petrol"),
            Car("Audi Q5", "Luxury", 900.0, 5, "Petrol"),
            Car("Mahindra XUV500", "SUV", 600.0, 7, "Diesel"),
            Car("Mahindra XUV300", "SUV", 450.0, 5, "Diesel"),
            Car("Tata Harrier", "SUV", 550.0, 5, "Diesel"),
            Car("Tata Safari", "SUV", 650.0, 7, "Diesel")
        ]
        CarRentalSystem.users = [
            User("user1", "password1"),
            User("user2", "password2")
        ]

    @staticmethod
    def login():
        print("")
        username = input("Enter username: ")
        password = input("Enter password: ")

        CarRentalSystem.logged_in_user = CarRentalSystem.authenticate_user(username, password)

        if CarRentalSystem.logged_in_user:
            print("Login successful!")
            CarRentalSystem.main_menu()
        else:
            print("Invalid username or password. Please try again.")

    @staticmethod
    def authenticate_user(username, password):
        for user in CarRentalSystem.users:
            if user.get_username() == username and user.get_password() == password:
                return user
        return None

    @staticmethod
    def register():
        new_username = input("Enter a new username: ")
        new_password = input("Enter a new password: ")

        CarRentalSystem.users.append(User(new_username, new_password))
        print("Registration successful. You can now log in.")

    @staticmethod
    def main_menu():
        while True:
            print("=========")
            print("Main Menu")
            print("=========")
            print("1. Rent a Car")
            print("2. View Rental History")
            print("3. View Car Sales")
            print("4. Reserve a Car")
            print("5. Logout")
            choice = int(input("Select an option: "))

            if choice == 1:
                CarRentalSystem.rent_car()
            elif choice == 2:
                CarRentalSystem.view_rental_history()
            elif choice == 3:
                CarRentalSystem.view_random_car_sales()
            elif choice == 4:
                CarRentalSystem.reserve_car()
            elif choice == 5:
                CarRentalSystem.logged_in_user = None
                print("Logged out.")
                return
            else:
                print("Invalid choice. Please try again.")

    @staticmethod
    def rent_car():
        if CarRentalSystem.logged_in_user:
            print()
            print("Available Car Categories:")
            print("=====================================")
            print("1. SUV")
            print("2. Luxury")
            print("=====================================")
            category_choice = int(input("Select a car category (enter the number): "))
            available_cars = [car for car in CarRentalSystem.cars if car.get_category() == ("SUV" if category_choice == 1 else "Luxury")]

            if available_cars:
                print("Available Cars:")
                print("=====================================")
                print("{:<5} {:<20} {:<10}".format("No.", "Car Name", "Daily Rate"))
                print("=====================================")

                for i, car in enumerate(available_cars, start=1):
                    car_info = "{:<5} {:<20} â‚¹{:<10.2f}".format(i, car.get_name(), car.get_daily_rate())
                    print(car_info)

                print("=====================================")
                car_choice = int(input("Select a car (enter the number): "))

                if 1 <= car_choice <= len(available_cars):
                    selected_car = available_cars[car_choice - 1]

                    rental_days = int(input("Enter rental days: "))
                    total_amount = selected_car.get_daily_rate() * rental_days

                    rental_record = Reservation(CarRentalSystem.logged_in_user, selected_car, None, rental_days)
                    CarRentalSystem.reservations.append(rental_record)

                    print("Reservation successful!")
                    print(f"Total amount: â‚¹{total_amount:.2f}/-")
                else:
                    print("Invalid car selection.")
            else:
                print("No cars available in the selected category.")
        else:
            print("You must log in to rent a car.")

    @staticmethod
    def reserve_car():
        if CarRentalSystem.logged_in_user:
            print()
            print("Available Car Categories:")
            print("=====================================")
            print("1. SUV")
            print("2. Luxury")
            print("=====================================")
            category_choice = int(input("Select a car category (enter the number): "))
            available_cars = [car for car in CarRentalSystem.cars if car.get_category() == ("SUV" if category_choice == 1 else "Luxury")]

            if available_cars:
                print("Available Cars:")
                print("=====================================")
                print("{:<5} {:<20} {:<10}".format("No.", "Car Name", "Daily Rate"))
                print("=====================================")

                for i, car in enumerate(available_cars, start=1):
                    car_info = "{:<5} {:<20} â‚¹{:<10.2f}".format(i, car.get_name(), car.get_daily_rate())
                    print(car_info)

                print("=====================================")
                car_choice = int(input("Select a car (enter the number): "))

                if 1 <= car_choice <= len(available_cars):
                    selected_car = available_cars[car_choice - 1]

                    rental_days = int(input("Enter rental days: "))
                    reservation_date = input("Enter reservation date (YYYY-MM-DD): ")

                    reservation_record = Reservation(CarRentalSystem.logged_in_user, selected_car, reservation_date, rental_days)
                    CarRentalSystem.reservations.append(reservation_record)

                    print("Reservation successful!")
                    print(f"Your reservation for {selected_car.get_name()} is confirmed for {rental_days} days starting from {reservation_date}.")
                else:
                    print("Invalid car selection.")
            else:
                print("No cars available in the selected category.")
        else:
            print("You must log in to reserve a car.")

    @staticmethod
    def view_rental_history():
        if CarRentalSystem.logged_in_user:
            print("=====================================")
            print(f"Rental History for {CarRentalSystem.logged_in_user.get_username()}:")
            print("=====================================")
            
            for record in CarRentalSystem.reservations:
                if record.get_user() == CarRentalSystem.logged_in_user:
                    print("Car:", record.get_car().get_name())
                    print("Reservation Date:", record.get_reservation_date())
                    print("Rental Days:", record.get_rental_days())
                    print("Total Amount: â‚¹{:.2f}".format(record.get_total_amount()))
                    print("------------------------")

            car_names = [record.get_car().get_name() for record in CarRentalSystem.reservations if
                         record.get_user() == CarRentalSystem.logged_in_user]
            rental_days = [record.get_rental_days() for record in CarRentalSystem.reservations if
                           record.get_user() == CarRentalSystem.logged_in_user]

            plt.bar(car_names, rental_days)
            plt.xlabel('Car Names')
            plt.ylabel('Rental Days')
            plt.title('Rental History')
            plt.show()
        else:
            print("You must log in to view rental history.")
            
    @staticmethod
    def view_random_car_sales():
        years = range(2020, 2025)
        random_sales_data = {year: random.randint(50, 200) for year in years}

        plt.plot(list(random_sales_data.keys()), list(random_sales_data.values()), marker='o', linestyle='-')
        plt.xlabel('Year')
        plt.ylabel('Number of Cars Sold')
        plt.title('Car Sales Data')
        plt.show()


if __name__ == "__main__":
    CarRentalSystem.main()
```
